### Description

In this small project i tried to train the YOLOv8n model on my custom dataset of axe to learn to detect it *(it can be trained to find whatever we want to*).


I prepared and annotated dataset of axes before train the model. And created in the Google Drive folder named YOLOv8CustomDatasetTrain. In the 'data' I create 2 new folders ('***images***' and '***labels***') and put inside prepared images and labels in special .txt format for YOLO network.

### Training the model

```python
import os
from ultralytics import YOLO

# Load a model
model = YOLO("yolov8n.pt")  

# Train the model
results = model.train(data=os.path.join(ROOT_DIR, "GoogleColab_config.yaml"), epochs=50)
```

### Results

```python
import cv2
import matplotlib.pyplot as plt

# Displaying results of training
results = cv2.imread('/content/gdrive/MyDrive/YOLOv8CustomDatasetTrain/runs/detect/train6/results.png')

# Convert BGR image to RGB
results = cv2.cvtColor(results, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 15))
plt.imshow(results)
plt.show()
```

```python
# Displaying some detecting results
detect = cv2.imread('/content/gdrive/MyDrive/YOLOv8CustomDatasetTrain/runs/detect/train6/val_batch0_labels.jpg')

# Convert BGR image to RGB
detect = cv2.cvtColor(detect, cv2.COLOR_BGR2RGB)
plt.figure(figsize=(15, 10))
plt.imshow(detect)
plt.show()
```

### **Conclusions**

As we can see, an axe was found in some images, but not at all, which indicates that the neural network should be additionally trained by applying hyperparameterization, data augmentation (***or any other technique that can be applied to improve the accuracy of prediction and detection to our model***)  , and be sure to increase the volume of the dataset on which we train the network, then the results will become better.
